```
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```
IPL = "Indian Premier League"
tweets = api.search(IPL, count=10, lang='en', exclude='retweets',
                    tweet_mode='extended')
for item in tweets:
    print(item)
    print("\n-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------\n")
```

# Fetching tweets based on hashtag and date and getting the users and their location

```
ipl2020 = "#IPL2020"
whistlepodu = "#whistlepodu"
date_since = "2020-09-18"
```

## #IPL2020 tweets

```
IPL2020tweets = tweepy.Cursor(api.search,
              q=ipl2020,
              lang="en",
              since=date_since).items(100)

iplfans_locs = [[tweet.user.screen_name, tweet.user.location] for tweet in IPL2020tweets]
iplfans_locs
```

```
ipl2020_tweet_text = pd.DataFrame(data=iplfans_locs, 
                    columns=['user', "location"])
ipl2020_tweet_text
```

## #whistlepodu tweets


```
whistlepodutweets = tweepy.Cursor(api.search,
              q=whistlepodu,
              lang="en",
              since=date_since).items(100)

cskfans_locs = [[tweet.user.screen_name, tweet.user.location] for tweet in whistlepodutweets]
cskfans_locs
```

```
whistlepodu_tweet_text = pd.DataFrame(data=cskfans_locs, 
                    columns=['user', "location"])
whistlepodu_tweet_text
```
